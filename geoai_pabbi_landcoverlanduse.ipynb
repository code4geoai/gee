This is the Unet implementation using the Pabbi Satellite image for land cover land use classification.

```python
import geoai
```

```python
train_raster_path = "./UnetV2/SourceV280.tif"
train_label_path = "./UnetV2/labelV280.tif"
test_raster_path = "./UnetV2/SourceV220.tif"
test_label_path = "./UnetV2/labelV220.tif"
```

```python

geoai.view_raster(test_label_path)
```

```python
out_folder = 'landcover_pabbi'
tiles = geoai.export_geotiff_tiles(
    in_raster = train_raster_path,
    out_folder= out_folder,
    in_class_data= train_label_path,
    class_value_field= 'labels',
    tile_size=128,
    stride=64,
    buffer_radius=0,
    skip_empty_tiles=True,
)
```

```python
# Train U-Net model
geoai.train_segmentation_model(
    images_dir=f"{out_folder}/images",
    labels_dir=f"{out_folder}/labels",
    output_dir=f"{out_folder}/unet_models",
    architecture="unet",
    encoder_name="resnet34",
    encoder_weights="imagenet",
    num_channels=4,
    num_classes=13,
    batch_size=8,
    num_epochs=50,
    learning_rate=0.001,
    val_split=0.2,
    verbose=True,
    plot_curves=True,
)
```
