```python
import leafmap
from geoai.download import (
    download_naip,
    download_overture_buildings,
    extract_building_stats,
)
```

```python
# define the path of the raster and vector.
raster_url = (
    "https://huggingface.co/datasets/giswqs/geospatial/resolve/main/naip_train.tif"
)
vector_url = "https://huggingface.co/datasets/giswqs/geospatial/resolve/main/naip_train_buildings.geojson"
```

```python
# lets download the data set and save it locally.

raster_path = geoai.download_file(raster_url)
vector_path = geoai.download_file(vector_url)
```

```python
#Printing raster information
geoai.print_raster_info(raster_path)
```

```python
# This return result as dictionary to be passed to other functions.
geoai.get_raster_info(raster_path)
```

```python
# Printing Vector informaiton
geoai.print_vector_info(vector_path)
```

```python
# Analyzing a specific column in the vector file.
geoai.analyze_vector_attributes(vector_path, "height")
```

```python
geoai.visualize_vector_by_attribute(
    vector_path,'height',)
```

```python
geoai.view_image(raster_path, figsize=(18,10))
```
